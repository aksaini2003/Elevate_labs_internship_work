```python
#first get all the unique movie_type
df['type'].head(20)
df['type'].value_counts().index
#but their is no need because we are not currently training the model
#if we have to train the model then 
# we have to convert all the categories and text into the numbers
```

```python
# like as follows 
d={'Movie':0,'TV Show':1}
# then do mapping or labelling
df['type']=df['type'].map(d)
```

```python
df.head()
```

```python
#lets do the same with the another columns that have categories 
#lets extract the unique country names 
unique_country=list()
for i in df['country'].values:
    for j in i:
        if j not in unique_country:
            unique_country.append(j)
```

```python
len(unique_country) #113 countries lets convert them in to the labels 
```

```python
countries=dict()
for i,c in enumerate(unique_country):
    countries[c]=i 
```

```python
countries #successfully given label to countries 
```

```python
#lets label the countries

def func(x):
    temp=list()
    for i in x:
        temp.append(countries[i])
    return temp
```

```python
df['country']=df['country'].map(func)
```

```python
df['country'] #successfully mapped 
```
